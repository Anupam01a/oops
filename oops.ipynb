8. How is encapsulation achieved in Python?
 - Encapsulation is the process of hiding data and restricting direct access to it.
 - Encapsulation achieved in Python by:-public,protected,private method
 - PUBLIC METHOD:- Accessible from anywhere from outside and inside of class.
 - PROTECTED METHOD:- Accessible within the class and its subclasses. Indicated by a single underscore(_var).
 - PRIVATE METHOD:-Accessible only inside the class. Indicated by double underscore(__var).


```python
class student:
  def __init__(self,name,marks):
    self.name = name     #public
    self._marks = marks  #protected
    self.__grade = 'A'   #private
  #getter for private variable
  def get_grade(self):
    return self.__grade
  #setter for private variable
  def set_grade(self,g):
    self.__grade = g
s=student("ram",90)
print(s.name)         #PUBLIC
print(s._marks)       #PROTECTED
print(s.get_grade())  #PRIVATE
```

 10. What are class and static methods in Python?
  - CLASS METHOD:-It is a special function that is bound with the class not the object. It can access or modify class variable but cannot instance variable.
  - It is defined using the @classmethod decorator.The first parameter is cls,which represents the class.

```python
class student:
  school_name = "VLM"

  @classmethod
  def change_school(cls,name):
    cls.school_name = name

student.change_school("PWSKILLS")
print(student.school_name)
```

 - It is defined using the @staticmethod decorator.


```python
class Math:

  @staticmethod
  def add(a,b):
    return a+b

print(Math.add(7,8))
```

12. What is method overriding in OOP?
 -  When a child class makes its own version of a method that already exists in the parent class. The child's method replaces(overrides) the parent's method when you call it using the child object.

```python
class Bank:
  def rate_of_interest(self):
    return "General bank interest is 5%"

class SBI(Bank):
  def rate_of_interest(self):       #method overridding
    return "SBI interest is 6%"

class HDFC(Bank):
  def rate_of_interest(self):       #method overridding
    return "HDFC interest is 7%"

B1=Bank()
S1=SBI()
H1=HDFC()
print(B1.rate_of_interest())
print(S1.rate_of_interest())
print(H1.rate_of_interest())
```

13. What is a property decorator in Python?
 -  A property decorator in Python allows you to use a method like a normal variable.It help you read or change a value safely without calling the method with parentheses. It is mainly used to control how a class attribute is accessed or updated.
 - Syntax:- @property

```python
class student:
  def __init__(self,marks):
    self._marks = marks

  @property             #property decorator
  def marks(self):
    return self._marks

obj = student(90)
print(obj.marks)        #no brackets
```

15. What is an abstract class in Python?
 -  Abstract class is used to hide complex implementation details.Abstract class may contain methods that are just declared and not implemented in abstract class itself. Subclass are responsible for implementing these abstract methods.
 - Abstract classes are created using the abc module and the @abstractmethod decorator.


```python
from abc import ABC, abstractmethod

class vehicle(ABC):

  #Abstract method declared but not implemented
  @abstractmethod
  def start(self):
    pass

class car(vehicle):
  def start(self):
    return "car starts with a key"

class bike(vehicle):
  def start(self):
    return "bike starts with a self-start button"

class electricscooter(vehicle):
  def start(self):
    return "electric scooter starts with a power button"

#objects of subclasses
C1 = car()
B1 = bike()
E1 = electricscooter()

print(C1.start())
print(B1.start())
print(E1.start())
```

18. What is multiple inheritance in Python?
 -  In multiple inheritance one child class has two or more parents class.
 -  This allows the child class to use the properties of all parents class.


```python
class father:
  def father_skill(self):
    return "Driving"

class mother:
  def mother_skill(self):
    return "Cooking"

class child(father,mother):     #child class inheriting from father and mother
  def child_skill(self):
    return "Studying"

C = child()
print(C.father_skill())    #from father
print(C.mother_skill())    #from mother
print(C.child_skill())     #from child
```

```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius

class rectangle(Shape):
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

C1 = circle(10)
R1 = rectangle(10,20)
print(C1.area())
print(R1.area())
```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
  def __init__(self,balance):
    self.__balance = balance

  def deposit(self,amount):
    self.__balance += amount
    print("Deposit successful.")

  def withdraw(self,amount):
    if self.__balance>=amount:
      self.__balance-=amount

      print("Withdraw successful.")

    else:
      print("Insufficient balance.")

  def get_balance(self):
    return self.__balance

acc1 = BankAccount(2000)
print("Account balance is",acc1.get_balance())
acc1.deposit(500)
print("Account balance is",acc1.get_balance())
acc1.withdraw(1000)
print("Account balance is",acc1.get_balance())



```

```python
""""14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary."""


class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus


# Creating object of Manager class
m1 = Manager(40, 500, 5000)

# Calculating and printing salary
print("Total Salary:", m1.calculate_salary())
```

```python
"""16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method."""


from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass


# Derived class Cow
class Cow(Animal):
    def sound(self):
        print("Cow says Moo")


# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        print("Sheep says Baa")


# Creating objects
c = Cow()
s = Sheep()

# Calling sound methods
c.sound()
s.sound()


```
